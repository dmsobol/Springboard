## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
pd.json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
pd.json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
worldBank_df.head()
codes = pd.DataFrame(worldBank_df['theme_namecode'])
```

```python
codes_no_na = codes.fillna(value={'code':'00','name':'Garbage'})
print(codes_no_na)
```

```python
worldBank_df.mjtheme_namecode
```

```python
worldBank_df.mjtheme_namecode[100]
```

```python
themes = worldBank_df.mjtheme_namecode.values
themes
```

```python
theme_list = [theme_dict for sublist in themes for theme_dict in sublist]
theme_list
```

```python
themes_df = pd.DataFrame(theme_list)
themes_df.head()
```

```python
themes_df.name.value_counts()
```

# 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# create dictionary of known values
code_dict = {}
for index, row in themes_df.iterrows():
    # if not in dict and not empty, add
    if (row['code']!= '' and row['name']!='') and (row['code'] not in code_dict.keys()):
        code_dict[row['code']]= row['name']
```

```python
code_dict
```

```python
# Create a dataframe with the missing names filled in
names_filled_df = themes_df
for index, row in names_filled_df.iterrows():
    if row['name']=='':
        row['name'] = code_dict[row['code']]
names_filled_df
```

```python
names_filled_df.name.value_counts()
```
